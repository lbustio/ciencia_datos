Edad por clase

Se compararán tres distribuciones de edad, agrupadas por clase del pasajero.
Como visualizaciones se utilizarán 2 gráficos:

1- `boxplot` + `stripplot` como antes.
2- `kdeplot`, para trazar curvas de densidad de edad para cada clase. Este método no puede manejar valores faltantes, por lo que se filtra los datos antes de utilizarlo.

```python
# set size
plt.figure(figsize=(20, 6))

# set palette
palette = sns.cubehelix_palette(5, start=3)

plt.subplot(1, 2, 1)
sns.boxplot(x='Pclass', y='Age', data=all_data.reset_index(),
            palette=palette, fliersize=0)

sns.stripplot(x='Pclass', y='Age', data=all_data.reset_index(),
              linewidth=0.6, palette=palette)
plt.xticks(np.arange(3), ['1st class', '2nd class', '3rd class'])
plt.title('Distribución de edad agrupada por clase de ticket (todos los datos)', fontsize=16)
plt.xlabel('Clase de ticket')

plt.subplot(1, 2, 2)

# Crear variables con datos filtrados para cada categoría
age_1_class = all_data[(all_data["Age"] > 0) &
                       (all_data["Pclass"] == 1)].reset_index()
age_2_class = all_data[(all_data["Age"] > 0) &
                       (all_data["Pclass"] == 2)].reset_index()
age_3_class = all_data[(all_data["Age"] > 0) &
                       (all_data["Pclass"] == 3)].reset_index()

# Graficar las tres variables creadas
sns.kdeplot(age_1_class["Age"], shade=True, color='#eed4d0', label='1st class')
sns.kdeplot(age_2_class["Age"], shade=True, color='#cda0aa', label='2nd class')
sns.kdeplot(age_3_class["Age"], shade=True, color='#a2708e', label='3rd class')
plt.title('Distribución de edad agrupada por clase de ticket (todos los datos)', fontsize=16)
plt.xlabel('Edad')
plt.xlim(0, 90)
plt.tight_layout()
plt.show()
```

##### ¿Qué hay en el nombre?

Cada valor del nombre del pasajero contiene el título del pasajero, que podemos extraer y descubrir. Luego se crea una nueva variable llamada "Título":

1. Se utiliza el método 'split' mediante una coma para dividir el nombre en dos partes y se guarda la segunda parte.
2. Se divide la parte guardada mediante un punto y se guarda la primera parte del resultado.
3. Para eliminar los espacios alrededor del título se utiliza el método 'split'.

Para visualizar cuántos pasajeros tienen cada título, se utiliza un gráfico de conteo (countplot).

```python
all_data['Title'] = all_data['Name'].str.split(',', expand = True)[1].str.split('.', expand = True)[0].str.strip(' ')

plt.figure(figsize=(6, 5))
ax = sns.countplot( x = 'Title', data = all_data, palette = "hls", order = all_data['Title'].value_counts().index)
_ = plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'  
)

plt.title('Passengers distribution by titles',fontsize= 14)
plt.ylabel('Number of passengers')

# calculate passengers for each category
labels = (all_data['Title'].value_counts())
# add result numbers on barchart
for i, v in enumerate(labels):
    ax.text(i, v+10, str(v), horizontalalignment = 'center', size = 10, color = 'black')
    

plt.tight_layout()
plt.show()
```

El segundo título más frecuente es **Miss** (señorita) - una mujer soltera, y el tercero es **Mrs**. (mujer casada). **Dr**. - Doctor es un título académico.

**Rev**. - Reverendo, es generalmente un título de cortesía para ministros o pastores cristianos protestantes.

A partir del número de la cabina, podemos extraer la primera letra, ¡lo cual nos dirá la ubicación de la cabina en el barco! Y me parece un conocimiento muy importante:

Encontré algunas descripciones de cada cubierta del Titanic. Había 8 cubiertas: la cubierta superior para los botes salvavidas, y las otras 7 se encontraban debajo y tenían símbolos de letras:

- **A**: no recorría toda la longitud del barco (es decir, no llegaba desde la popa hasta la proa del barco) y estaba destinada a pasajeros de 1ra clase.
- **B**: no recorría toda la longitud del barco (se interrumpía a 37 metros por encima de la cubierta C) y servía como lugar para los anclajes en la parte delantera.
- **C**: en la parte delantera se encontraba la cocina, el comedor de la tripulación, así como un área de paseo para pasajeros de 3ra clase.
- **D**: un área de paseo para pasajeros.
- **E**: cabinas de 1ra y 2da clase.
- **F**: parte de las cabinas de pasajeros de 2da clase y la mayoría de las cabinas de 3ra clase.
- **G**: no recorría toda la longitud del barco, las salas de calderas se encontraban en el centro.
- **T**: ¿cubierta de botes?

Para los pasajeros sin información sobre la cubierta, asignaré la letra **U** (desconocido).

#### Valores perdidos

Los valores que faltan pueden deberse a diferentes motivos; existen tres mecanismos principales:

- Missing Completely at Random (MCAR) - no hay relación entre la falta de datos y ningún valor, observado o faltante.
- Falta al azar (MAR): existe una relación sistemática entre la propensión de los valores que faltan y los datos observados, pero no los que faltan.
- Falta no aleatoria (MNAR): existe una relación entre la propensión de un valor a faltar y sus valores.

Dependiendo de las razones por las que faltan valores, debemos actuar de forma diferente con la imputación de datos. No es fácil decir con seguridad qué tipo de valores faltantes tenemos, pero podemos hacer algunas suposiciones.

Por ejemplo, basándonos en el hecho de que entre los pasajeros sin cabina especificada, el porcentaje de ahogados es mayor, podríamos pensar que la presencia de esta información nos dice algo (MAR). Falta información sobre la cabina de un número muy elevado de pasajeros (687 NA en el conjunto de datos de entrenamiento y 327 en el conjunto de datos de prueba). He leído sobre las suposiciones de que esta información se recogió de las palabras de los pasajeros supervivientes. #### Ingeniería de características

A partir de 11 variables, generaré 31 características para cada pasajero.
Tenemos que transformar las variables categóricas en números, porque la mayoría de los modelos sólo aceptan números como entrada. Utilizaré las siguientes formas de hacerlo:

1. Variables ficticias: una variable numérica utilizada para representar subgrupos de la muestra en su estudio. Toma sólo el valor 0 ó 1 para indicar la ausencia o presencia de alguna característica. Este método es muy adecuado para variables categóricas nominales, que no tienen un orden intrínseco en sus categorías.

2. Reemplazar valores de cadena por números - adecuado para variables ordinales (como la clase de billete de pasajero) o si sólo tenemos dos valores posibles (En este conjunto de datos podemos implementarlo a Sexo, por ejemplo).

Características:

- **Cubierta** (8 características) - Del número del camarote podemos extraer la primera letra, que nos indicará la ubicación del camarote en el barco (Cubierta). Si no hay número de Caín, el valor será **U** - desconocido. Utilizando la función get_dummies, lo convierto en 8 características ficticias con el prefijo **deck**.
- **Title_category** (8 características) - Ya hemos creado la variable **Title_category** en la parte de imputación de faltas. Utilizando la función get_dummies, la convierto en 8 variables ficticias con el prefijo **title**.
- **Family_size_group** (4 características) - Calculo **Family_size** resumiendo las variables **SibSp** y **Parch** y añadiendo 1. Luego creo **Family_size**. Luego creo **Family_size_group** (4 valores posibles) basado en el tamaño de la familia. Utilizando la función get_dummies, lo convierto en 8 características ficticias con el prefijo **family**.
- **Age_NA** (1 característica) - Creamos esta variable durante el proceso de imputación de faltas.
- Tarifa_cero** (1 característica) - Marca a las personas con tarifa cero, ya que pueden estar relacionadas con la White Star Line.
- Embarcado** (3 características) - Utilizando la función get_dummies, convierto la variable **Embarcado** en 3 características ficticias con el prefijo **embarcado**.
- Sexo (1 característica): como sólo tenemos 2 valores posibles, la convierto en una característica binaria: **1** si es mujer, **0** si es hombre.
- Edad (1 característica): ya es numérica (continua).
- Fare (1 característica) - ya numérica (continua).
- Pclass (1 característica) - ya numérica (variable categórica ordinal).
- SibSp (1 característica): como ya utilizamos el tamaño de la familia, convierto **SibSp** en una característica binaria: **1** si hay hermanos / cónyuges a bordo del Titanic, **0** si no los hay.
- **Parch** (1 característica) - como ya usamos el tamaño de la familia, convierto **Parch** en una característica binaria - **1** si hay padres / hijos a bordo del Titanic, **0** si no.

```python
def feature_generator (data, train = False):
    
    features_data = data
    
    # Deck
    # Extract deck letter from cabin number
    features_data['deck'] = features_data['Cabin'].str.split('', expand = True)[1]
    # If cabin is NA - deck = U
    features_data.loc[features_data['deck'].isna(), 'deck'] = 'U'
    # If cabin is T - change to A (see EDA)
    features_data.loc[features_data['deck'] == 'T', 'deck'] = 'A'
    # Create dummy variables with prefix 'deck'
    features_data = pd.concat([features_data,
                               pd.get_dummies(features_data['deck'], prefix = 'deck')], 
                               axis=1)
    
    
    # titles dummy
    features_data = pd.concat([features_data, 
                               pd.get_dummies(features_data['Title_category'],
                                              prefix = 'title')], axis=1)

    # family size
    features_data['Family_size'] = features_data['SibSp'] + features_data['Parch'] + 1
    features_data['Family_size_group'] = features_data['Family_size'].map(
                                            lambda x: 'f_single' if x == 1 
                                                    else ('f_usual' if 5 > x >= 2 
                                                          else ('f_big' if 8 > x >= 5 
                                                               else 'f_large' )))
    features_data = pd.concat([features_data, 
                               pd.get_dummies(features_data['Family_size_group'], 
                                              prefix = 'family')], axis=1)     
    
    
    # Sex to number
    features_data['Sex'] = features_data['Sex'].map({'female': 1, 'male': 0}).astype(int)
    
    # embarked dummy
    features_data = pd.concat([features_data, 
                               pd.get_dummies(features_data['Embarked'], 
                                              prefix = 'embarked')], axis=1)
    
    # zero fare feature
    features_data['zero_fare'] = features_data['Fare'].map(lambda x: 1 if x == 0 else (0))
    
    # from numeric to categorical
    features_data['SibSp'] = features_data['SibSp'].map(lambda x: 1 if x > 0 else (0))
    features_data['Parch'] = features_data['Parch'].map(lambda x: 1 if x > 0 else (0))
    
    # delete variables we are not going to use anymore
    del features_data['PassengerId']
    del features_data['Ticket']
    del features_data['Cabin']
    del features_data['deck']    
    del features_data['Title_category']
    del features_data['Name']
    del features_data['Family_size']
    del features_data['Family_size_group'] 
    del features_data['Embarked']    
    
    return features_data
```

```python
# Generate features from training dataset
all_train_features = feature_generator(train_data)
# Generate features from test dataset
all_test_features = feature_generator(test_data)
```

```python
# Generar datos aleatorios
datos = []
fecha_actual = datetime.now().date()
for _ in range(100):
	producto = random.choice(productos)
	categoria = random.choice(categorias)
	fecha = fecha_actual - timedelta(days=random.randint(0, 365))
	cantidad = random.randint(1, 20)
	precio = round(random.uniform(10, 50), 2)
	datos.append([producto, categoria, fecha, cantidad, precio])
```

```python
# Crear DataFrame y guardar en archivo CSV
df = pd.DataFrame(datos, columns=["Producto", "Categoria", "Fecha", 
"Cantidad", "Precio"])
df.to_csv("datos_ventas.csv", index=False)
```

### Solución de la tarea

```python
# Importar las librerias que se usarán
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

a)	Cargar los datos desde el archivo CSV y almacenarlos en un DataFrame llamado ventas.

```python
# Paso 1: Cargar los datos desde el archivo CSV y almacenarlos en un DataFrame llamado "ventas"
ventas = pd.read_csv('data/10.-datos_ventas.csv')
```

b)	Calcular y mostrar el resumen estadístico completo del DataFrame, incluyendo medidas de tendencia central, dispersión y forma.

```python
# Paso 2: Calcular y mostrar el resumen estadístico completo del DataFrame
resumen_estadistico = ventas.describe()
print(resumen_estadistico)
```

c)	Realizar un análisis por categoría de productos. Para cada categoría, calcular y mostrar:

```python
# Paso 3: Realizar un análisis por categoría de productos
analisis_categorias = ventas.groupby('Categoria').agg({
    'Cantidad': 'sum',
    'Precio': ['mean', 'max', 'min'],
})
print(analisis_categorias)
```

d)	Calcular y mostrar la matriz de correlación entre las columnas “cantidad” y “precio”, así como también visualizarla mediante un mapa de calor.

```python
# Paso 4: Calcular y mostrar la matriz de correlación entre las columnas "cantidad" y "precio"
matriz_correlacion = ventas[['Cantidad', 'Precio']].corr()
print(matriz_correlacion)

# Visualización de la matriz de correlación mediante un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

e)	Identificar los productos con ventas atípicas (outliers) en función de la cantidad vendida. Mostrar los nombres de los productos y las cantidades correspondientes.

```python
# Paso 5: Identificar los productos con ventas atípicas (outliers) en función de la cantidad vendida
q1 = ventas['Cantidad'].quantile(0.25)
q3 = ventas['Cantidad'].quantile(0.75)
iqr = q3 - q1
umbral_superior = q3 + 1.5 * iqr
outliers_cantidad = ventas[ventas['Cantidad'] > umbral_superior][['Producto', 'Cantidad']]
print(outliers_cantidad)
```

f)	Calcular y mostrar el porcentaje de ventas acumuladas por categoría. Ordenar las categorías de mayor a menor por su porcentaje de ventas acumuladas.

```python
# Paso 6: Calcular y mostrar el porcentaje de ventas acumuladas por categoría
ventas_por_categoria = ventas.groupby('Categoria')['Cantidad'].sum()
porcentaje_ventas_acumuladas = ventas_por_categoria / ventas['Cantidad'].sum() * 100
porcentaje_ventas_acumuladas = porcentaje_ventas_acumuladas.sort_values(ascending=False)
print(porcentaje_ventas_acumuladas)
```

g)	Realizar un análisis temporal de las ventas. 

```python
# Paso 7: Realizar un análisis temporal de las ventas
ventas['mes'] = pd.to_datetime(ventas['Fecha']).dt.month
analisis_temporal = ventas.groupby('mes').agg({
    'Cantidad': 'sum',
    'Precio': 'mean',
})
mes_mayor_monto_ventas = analisis_temporal['Precio'].idxmax()
mes_menor_monto_ventas = analisis_temporal['Precio'].idxmin()
print(analisis_temporal)
print("Mes con el mayor monto total de ventas:", mes_mayor_monto_ventas)
print("Mes con el menor monto total de ventas:", mes_menor_monto_ventas)
```

h)	Exportar los resultados obtenidos en los pasos 3, 4, 5, 6 y 7 a un nuevo archivo CSV llamado “resultados_ventas.csv”.

```python
# Paso 8: Exportar los resultados obtenidos a un nuevo archivo CSV llamado "resultados_ventas.csv"
resumen_estadistico.to_csv('resultados_ventas.csv', index=False, header=True)
analisis_categorias.to_csv('resultados_ventas.csv', mode='a', index=True, header=True)
matriz_correlacion.to_csv('resultados_ventas.csv', mode='a', index=True, header=True)
outliers_cantidad.to_csv('resultados_ventas.csv', mode='a', index=False, header=True)
porcentaje_ventas_acumuladas.to_csv('resultados_ventas.csv', mode='a', index=True, header=True)
analisis_temporal.to_csv('resultados_ventas.csv', mode='a', index=True, header=True)
```
