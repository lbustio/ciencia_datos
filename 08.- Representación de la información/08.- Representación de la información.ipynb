### Gráficos

#### Area

Un gráfico de áreas es un tipo de gráfico que representa datos cuantitativos utilizando áreas coloreadas o sombreadas debajo de una línea o curva. Está compuesto por una serie de puntos conectados por segmentos de línea recta y el área entre la línea y el eje x se llena de color.

El gráfico de áreas se utiliza para mostrar la distribución o la comparación de datos a lo largo de un eje x común. Algunos casos de uso comunes son:

1. Comparación de datos: Permite visualizar la relación relativa entre dos o más series de datos. Se puede utilizar para comparar cambios en el tiempo o entre diferentes categorías.

2. Tendencias: Ayuda a identificar patrones o tendencias en los datos a lo largo de un período o categoría.

3. Análisis de acumulación: Permite analizar la contribución de cada serie de datos a medida que se acumula a lo largo del eje x.

Ventajas de los gráficos de áreas:
- Proporcionan una visualización clara de la relación relativa entre varias series de datos.
- Permiten identificar patrones y tendencias en los datos de forma rápida.
- Son útiles para comparar cambios a lo largo del tiempo o entre categorías.

Desventajas de los gráficos de áreas:
- Pueden volverse visualmente confusos o difíciles de interpretar si se superponen muchas áreas.
- La representación del área puede ser susceptible a errores perceptuales, ya que el enfoque está en la comparación de áreas y no en las longitudes absolutas.

En general, los gráficos de áreas son útiles para resaltar las relaciones y las diferencias entre las series de datos. Sin embargo, es importante tener en cuenta las limitaciones y considerar el contexto y la claridad de la información que se desea comunicar al seleccionar este tipo de gráfico.

```python
# Area
import numpy as np # libreria para el trabajo con matrices
import matplotlib.pyplot as plt # libreria para la visualizacion
import random # libreria par el manejo de numeros aleatorios

rng = 10
x = range(0, rng)
y = list() # lluvia en el 2019
z = list() # lluvia en el 2020

for i in range(rng):
  y.append(random.randint(0, 9))
  z.append(random.randint(0,15))

# Encabezados y personalizacion general del grafico
plt.title("Precipitaciones por año")

# Dibujos de los graficos
plt.fill_between(x, z)
plt.fill_between(x, y)

# Personalizaciones finas
# Personalizando X
plt.xlabel("x label")
plt.tick_params(axis='x', colors='red', direction='out', length=7, width=2)
plt.xticks(x, ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sept", "Oct"])

# Personalizando Y
plt.ylabel("y label")

# Mostrar el resultado
plt.show()
```

#### Barplot

Un barplot, también conocido como gráfico de barras, es un tipo de gráfico que representa datos categóricos o discretos utilizando barras rectangulares de longitud proporcional a los valores que se están representando. Cada barra representa una categoría y su altura representa la magnitud o frecuencia de esa categoría.

El barplot se utiliza para mostrar la comparación entre diferentes categorías o grupos de datos. Algunos casos de uso comunes son:

1. Comparación de valores: Permite visualizar rápidamente las diferencias en los valores entre categorías o grupos.

2. Análisis de tendencias: Ayuda a identificar patrones o tendencias en los datos a lo largo de diferentes categorías.

3. Presentación de resultados: Es útil para presentar de manera clara y concisa información cuantitativa o cualitativa a un público.

Ventajas de los barplots:
- Son fáciles de entender y interpretar, ya que las diferencias en las longitudes de las barras son intuitivas para la comparación visual.
- Permiten mostrar grandes cantidades de datos en un formato claro y conciso.
- Son versátiles y se pueden personalizar para resaltar aspectos específicos de los datos, como agregar colores, etiquetas, etc.

Desventajas de los barplots:
- Si hay muchas categorías o barras, el gráfico puede volverse desordenado y difícil de leer.
- Al representar valores continuos con barras, se pierde cierta información detallada sobre los datos.

En general, los barplots son una herramienta eficaz para la comparación y visualización de datos categóricos o discretos. Son especialmente útiles cuando se necesita resaltar las diferencias o patrones entre categorías o grupos de datos. Sin embargo, se debe tener cuidado al interpretar los datos y considerar las limitaciones del gráfico en relación con el tipo de datos que se están representando.

```python
# Barplot

# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Crear un conjunto de datos aleatorio
height = [3, 12, 5, 18, 45]  # Altura de las barras
bars = ('A', 'B', 'C', 'D', 'E')  # Etiquetas de las barras
y_pos = np.arange(len(bars))  # Posiciones en el eje y

# Crear las barras
plt.bar(y_pos, height)

# Asignar etiquetas en el eje x
plt.xticks(y_pos, bars)

# Mostrar el gráfico
plt.show()
```

#### Boxplot

Un boxplot, también conocido como diagrama de caja y bigotes, es un tipo de gráfico que muestra la distribución de un conjunto de datos numéricos a través de cinco estadísticas principales: el valor mínimo, el primer cuartil (Q1), la mediana (Q2 o valor medio), el tercer cuartil (Q3) y el valor máximo.

El boxplot se utiliza para resumir y visualizar la distribución de los datos, proporcionando información sobre la simetría, los valores atípicos y la dispersión del conjunto de datos. Algunos casos de uso comunes son:

1. Comparación de distribuciones: Permite comparar fácilmente la forma y dispersión de múltiples distribuciones de datos.

2. Detección de valores atípicos: Ayuda a identificar valores extremos o atípicos que pueden estar presentes en los datos.

3. Análisis de dispersión: Proporciona información sobre la variabilidad y la dispersión de los datos al mostrar los rangos intercuartílicos y los bigotes.

Ventajas de los boxplots:
- Proporcionan una representación visual compacta de la distribución de los datos y resumen las estadísticas clave.
- Permiten identificar valores atípicos y entender la simetría de los datos.
- Son útiles para comparar distribuciones entre diferentes categorías o grupos.

Desventajas de los boxplots:
- No proporcionan información detallada sobre la forma exacta de la distribución, ya que se simplifica en cinco estadísticas principales.
- No son adecuados para datos categóricos o discretos, ya que se centran en la distribución de valores numéricos.

En general, los boxplots son una herramienta poderosa para resumir y comparar distribuciones de datos numéricos. Son especialmente útiles cuando se necesita comprender la dispersión, los valores atípicos y la forma general de los datos. Sin embargo, es importante considerar el contexto de los datos y combinarlos con otras visualizaciones o análisis para obtener una comprensión completa del conjunto de datos.

#### Bubble graph

Un bubble graph, también conocido como gráfico de burbujas, es un tipo de gráfico que representa datos utilizando burbujas o círculos. Cada burbuja tiene un tamaño proporcional a una variable numérica adicional, además de las coordenadas x e y que representan las variables principales.

El bubble graph se utiliza para mostrar relaciones entre tres variables de manera visualmente impactante. Algunos casos de uso comunes son:

1. Comparación de variables: Permite comparar dos variables principales (ejes x e y) y representar una tercera variable (tamaño de la burbuja) para mostrar una relación tridimensional.

2. Identificación de patrones: Ayuda a identificar patrones o agrupaciones de datos al observar la distribución y el tamaño de las burbujas.

3. Presentación de datos complejos: Es útil para representar conjuntos de datos que contienen múltiples dimensiones y variables.

Ventajas de los bubble graphs:
- Proporcionan una visualización atractiva y efectiva para representar relaciones entre tres variables.
- Permiten comparar datos en dos dimensiones principales y agregar una tercera dimensión mediante el tamaño de las burbujas.
- Son útiles para mostrar conjuntos de datos complejos o grandes de manera clara y concisa.

Desventajas de los bubble graphs:
- Pueden volverse confusos o difíciles de interpretar si hay demasiadas burbujas o si los tamaños varían mucho entre ellas.
- La representación tridimensional puede dificultar la comparación precisa entre las variables principales.

En general, los bubble graphs son una herramienta visual efectiva para representar relaciones complejas entre tres variables. Son especialmente útiles cuando se necesita resaltar y comparar datos en múltiples dimensiones. Sin embargo, es importante considerar el contexto y la legibilidad de los datos al utilizar este tipo de gráfico y asegurarse de que se utilice de manera adecuada y clara para el conjunto de datos en cuestión.

```python
# bubble graph

# Importar las librerías necesarias
import matplotlib.pyplot as plt
import seaborn as sns
from gapminder import gapminder  # Conjunto de datos

# Cargar el conjunto de datos
data = gapminder.loc[gapminder.year == 2007]  # Filtrar datos para el año 2007

# Utilizar la función scatterplot para construir el mapa de burbujas
sns.scatterplot(data=data, x="gdpPercap", y="lifeExp", size="pop", legend=False, sizes=(20, 2000))

# Mostrar el gráfico
plt.show()
```

#### Bubble Map

Un bubble map, también conocido como mapa de burbujas, es un tipo de visualización que combina un mapa geográfico con burbujas o círculos de diferentes tamaños para representar datos relacionados con ubicaciones geográficas específicas.

El bubble map se utiliza para mostrar la distribución geográfica de datos y resaltar diferencias o variaciones en función del tamaño de las burbujas. Algunos casos de uso comunes son:

1. Visualización de datos geográficos: Permite representar datos numéricos o categorías en un mapa para identificar patrones o concentraciones geográficas.

2. Análisis espacial: Ayuda a entender la distribución y la relación entre variables en diferentes ubicaciones geográficas.

3. Presentación de datos demográficos o estadísticos: Es útil para mostrar datos relacionados con poblaciones, ingresos, tasas de participación, etc., en diferentes regiones.

Ventajas del bubble map:
- Proporciona una representación visual efectiva de datos geográficos y permite una comprensión rápida de las diferencias en ubicaciones específicas.
- El tamaño de las burbujas permite comparar magnitudes o valores numéricos relacionados con cada ubicación.
- Facilita la identificación de patrones o tendencias espaciales en los datos.

Desventajas del bubble map:
- La representación de datos está limitada a ubicaciones geográficas específicas y no es adecuada para visualizar datos que no están relacionados con la geografía.
- Puede ser difícil de interpretar si hay superposición de burbujas o si los tamaños varían significativamente.

En resumen, el bubble map es una herramienta poderosa para visualizar y analizar datos geográficos y proporcionar una representación visual atractiva de patrones y diferencias en diferentes ubicaciones. Sin embargo, es importante considerar el contexto de los datos y utilizarlo adecuadamente para resaltar información relevante y comprensible en relación con la geografía y los datos específicos que se están visualizando.

```python
# Bubble map
# Importar la librería necesaria
import plotly.express as px

# Cargar el conjunto de datos y filtrar para el año 2007
df = px.data.gapminder().query("year==2007")

# Crear la figura del gráfico de dispersión geográfico
fig = px.scatter_geo(df, locations="iso_alpha", color="continent",
                     hover_name="country", size="pop",
                     projection="natural earth")

# Mostrar el gráfico
fig.show()
```

#### Bullet Graph

Un bullet graph es un tipo de gráfico que combina características de un gráfico de barras y un indicador de desempeño para representar una métrica o medida específica junto con metas y rangos de referencia.

El bullet graph se utiliza principalmente para mostrar el progreso o rendimiento de una métrica o indicador en relación con una meta establecida. Algunos casos de uso comunes son:

1. Seguimiento de metas: Permite visualizar fácilmente el progreso hacia una meta establecida y compararlo con rangos de referencia, como metas mensuales o anuales.

2. Evaluación de desempeño: Proporciona una representación visual clara de cómo un valor se compara con un rango objetivo o con los valores esperados.

3. Presentación de métricas clave: Es útil para mostrar métricas importantes de manera concisa y resaltar áreas de éxito o de mejora.

Ventajas del bullet graph:
- Proporciona una representación visual efectiva del rendimiento o progreso en relación con metas y rangos de referencia.
- Es compacto y eficiente en términos de espacio, lo que permite mostrar múltiples indicadores en un solo gráfico.
- Permite una rápida interpretación al presentar claramente las medidas de desempeño y las comparaciones con las metas establecidas.

Desventajas del bullet graph:
- No es adecuado para representar múltiples variables o datos complejos.
- Puede ser limitado para mostrar información adicional o detalles sobre la métrica en sí.

En general, el bullet graph es una herramienta eficaz para visualizar el progreso o rendimiento en relación con metas y rangos de referencia. Es especialmente útil cuando se necesita una representación visual clara y concisa de una métrica importante. Sin embargo, se debe tener cuidado al seleccionar y configurar adecuadamente las métricas, metas y rangos de referencia para asegurarse de que la visualización sea comprensible y significativa para el público objetivo.

```python
# Bullet graph

# Importar la librería necesaria
import plotly.graph_objects as go

# Crear una figura vacía
fig = go.Figure()

# Añadir el primer indicador (Revenue)
fig.add_trace(go.Indicator(
    mode = "number+gauge+delta", value = 180,  # Configurar el valor y el modo del indicador
    delta = {'reference': 200},  # Configurar la referencia para el delta
    domain = {'x': [0.25, 1], 'y': [0.08, 0.25]},  # Configurar la posición en el gráfico
    title = {'text': "Revenue"},  # Configurar el título del indicador
    gauge = {
        'shape': "bullet",  # Configurar la forma del indicador
        'axis': {'range': [None, 300]},  # Configurar el rango del eje
        'threshold': {
            'line': {'color': "black", 'width': 2},  # Configurar la línea del umbral
            'thickness': 0.75,  # Configurar el grosor del umbral
            'value': 170},  # Configurar el valor del umbral
        'steps': [
            {'range': [0, 150], 'color': "gray"},  # Configurar los rangos y colores de los pasos
            {'range': [150, 250], 'color': "lightgray"}],  # Configurar los rangos y colores de los pasos
        'bar': {'color': "black"}}))  # Configurar el color de la barra

# Añadir el segundo indicador (Profit)
fig.add_trace(go.Indicator(
    mode = "number+gauge+delta", value = 35,  # Configurar el valor y el modo del indicador
    delta = {'reference': 200},  # Configurar la referencia para el delta
    domain = {'x': [0.25, 1], 'y': [0.4, 0.6]},  # Configurar la posición en el gráfico
    title = {'text': "Profit"},  # Configurar el título del indicador
    gauge = {
        'shape': "bullet",  # Configurar la forma del indicador
        'axis': {'range': [None, 100]},  # Configurar el rango del eje
        'threshold': {
            'line': {'color': "black", 'width': 2},  # Configurar la línea del umbral
            'thickness': 0.75,  # Configurar el grosor del umbral
            'value': 50},  # Configurar el valor del umbral
        'steps': [
            {'range': [0, 25], 'color': "gray"},  # Configurar los rangos y colores de los pasos
            {'range': [25, 75], 'color': "lightgray"}],  # Configurar los rangos y colores de los pasos
        'bar': {'color': "black"}}))  # Configurar el color de la barra

# Añadir el tercer indicador (Satisfaction)
fig.add_trace(go.Indicator(
    mode = "number+gauge+delta", value = 220,  # Configurar el valor y el modo del indicador
    delta = {'reference': 200},  # Configurar la referencia para el delta
    domain = {'x': [0.25, 1], 'y': [0.7, 0.9]},  # Configurar la posición en el gráfico
    title = {'text' :"Satisfaction"},  # Configurar el título del indicador
    gauge = {
        'shape': "bullet",  # Configurar la forma del indicador
        'axis': {'range': [None, 300]},  # Configurar el rango del eje
        'threshold': {
            'line': {'color': "black", 'width': 2},  # Configurar la línea del umbral
            'thickness': 0.75,  # Configurar el grosor del umbral
            'value': 210},  # Configurar el valor del umbral
        'steps': [
            {'range': [0, 150], 'color': "gray"},  # Configurar los rangos y colores de los pasos
            {'range': [150, 250], 'color': "lightgray"}],  # Configurar los rangos y colores de los pasos
        'bar': {'color': "black"}}))  # Configurar el color de la barra

# Actualizar el diseño de la figura
fig.update_layout(height = 400 , margin = {'t':0, 'b':0, 'l':0})

# Mostrar el gráfico
fig.show()
```

#### Gráfico de acordes

Un chord graph, también conocido como gráfico de acordes o diagrama de cuerdas, es un tipo de visualización que muestra las relaciones entre diferentes categorías o grupos de datos. Está compuesto por círculos o arcos que representan las categorías, y las conexiones o cuerdas entre ellos indican las conexiones o relaciones entre las categorías.

El chord graph se utiliza para representar redes o relaciones complejas entre diferentes entidades o elementos. Algunos casos de uso comunes son:

1. Visualización de relaciones de red: Permite mostrar interacciones o conexiones entre diferentes nodos o elementos.

2. Análisis de flujos o transferencias: Ayuda a comprender la transferencia o movimiento de datos, recursos o entidades entre diferentes categorías o grupos.

3. Identificación de patrones o agrupaciones: Es útil para identificar grupos o comunidades de entidades relacionadas y sus conexiones.

Ventajas del chord graph:
- Proporciona una visualización intuitiva de las relaciones y conexiones entre diferentes categorías o grupos.
- Permite identificar patrones de flujo o transferencia de datos entre diferentes elementos.
- Es útil para mostrar redes complejas de manera concisa y comprensible.

Desventajas del chord graph:
- Puede volverse visualmente complejo y confuso si hay muchas categorías o conexiones.
- La interpretación precisa puede requerir un análisis más detallado y conocimiento del dominio.

En resumen, el chord graph es una herramienta poderosa para visualizar y analizar relaciones complejas entre diferentes categorías o grupos. Es especialmente útil cuando se necesita entender las conexiones y flujos de datos o entidades en un sistema o red. Sin embargo, se debe tener cuidado al seleccionar y configurar adecuadamente los elementos y las conexiones para asegurarse de que la visualización sea legible y útil para el propósito específico.

```python
import plotly.graph_objects as go
import networkx as nx
import numpy as np

# Crear los datos de la tabla
data = {
    'Emma': [16, 3, 28, 0, 18],
    'Isabella': [18, 0, 12, 5, 29],
    'Ava': [9, 11, 17, 27, 0],
    'Olivia': [19, 0, 31, 11, 12],
    'Sophia': [23, 17, 10, 0, 34]
}

# Crear un grafo de red
G = nx.DiGraph()

# Añadir nodos
for node in data:
    G.add_node(node)

# Añadir aristas con los pesos de los datos
for node, connections in data.items():
    for i, weight in enumerate(connections):
        target_node = list(data.keys())[i]
        if weight > 0:  # Solo agregar aristas con peso positivo
            G.add_edge(node, target_node, weight=weight)

# Generar posiciones de nodos para el diagrama de cordón
pos = nx.circular_layout(G)

# Extraer las aristas y sus pesos
edges = G.edges(data=True)
edge_x = []
edge_y = []
edge_text = []
for edge in edges:
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x.append(x0)
    edge_x.append(x1)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_text.append(edge[2]['weight'])

# Crear el gráfico de cordón con Plotly
fig = go.Figure()

# Añadir aristas al gráfico
fig.add_trace(go.Scatter(
    x=edge_x, y=edge_y,
    mode='lines+text',
    line=dict(width=2, color='rgba(0,0,0,0.5)'),
    text=edge_text,
    textposition='top center',
    hoverinfo='text'
))

# Añadir nodos al gráfico
node_x = [pos[node][0] for node in G.nodes()]
node_y = [pos[node][1] for node in G.nodes()]
node_text = list(G.nodes())

fig.add_trace(go.Scatter(
    x=node_x, y=node_y,
    mode='markers+text',
    text=node_text,
    textposition='top center',
    marker=dict(size=10, color='rgba(0,0,0,0.7)'),
    hoverinfo='text'
))

# Configurar el diseño del gráfico
fig.update_layout(
    showlegend=False,
    xaxis=dict(showgrid=False, zeroline=False),
    yaxis=dict(showgrid=False, zeroline=False),
    title_text='Chord Graph'
)

# Mostrar el gráfico
fig.show()
```

#### Candle plot

Un candle plot, también conocido como gráfico de velas o gráfico de candelabro, es un tipo de gráfico utilizado para representar la fluctuación de precios de un activo financiero en un período de tiempo determinado. Es comúnmente utilizado en el análisis técnico de los mercados financieros, como el mercado de valores o el mercado de divisas.

El candle plot se compone de barras verticales llamadas "cuerpos" y líneas horizontales llamadas "mechas" o "sombras". Cada barra o vela representa la información del precio de apertura, cierre, máximo y mínimo de un período de tiempo específico.

Algunos componentes clave del candle plot son:

1. Cuerpo de la vela: Representa la diferencia entre el precio de apertura y el precio de cierre del activo en el período de tiempo especificado. El cuerpo se muestra como una barra vertical que está coloreada o rellena de acuerdo con si el precio de cierre fue mayor o menor que el precio de apertura.

2. Mechas o sombras: Representan los precios máximo y mínimo alcanzados durante el período de tiempo. Se muestran como líneas verticales que se extienden desde el cuerpo de la vela hacia arriba y hacia abajo.

El candle plot se utiliza para proporcionar una representación visual de la acción del precio y la volatilidad en un mercado financiero. Algunos casos de uso comunes son:

1. Análisis de tendencias: Permite identificar patrones y tendencias alcistas o bajistas en los precios de un activo financiero.

2. Identificación de puntos de inflexión: Ayuda a identificar posibles niveles de soporte y resistencia en los precios.

3. Evaluación de la volatilidad: Permite evaluar la amplitud de los movimientos de precios durante un período determinado.

Ventajas del candle plot:

#### Errorbar Plot

Un errorbar plot, también conocido como gráfico de barras de error, es un tipo de visualización que representa la incertidumbre o variabilidad asociada a un conjunto de datos. Consiste en una barra vertical o horizontal que se extiende desde un punto central y representa el rango de error o incertidumbre en torno a ese punto.

El errorbar plot se utiliza para mostrar la dispersión de los datos y proporcionar información sobre la precisión o confiabilidad de las mediciones. Algunos casos de uso comunes son:

1. Representación de datos experimentales: Permite mostrar el error o la variabilidad asociada a las mediciones experimentales, lo que ayuda a interpretar la confiabilidad de los resultados.

2. Comparación de grupos o condiciones: Es útil para comparar la variabilidad entre diferentes grupos o condiciones y evaluar si hay diferencias significativas.

3. Presentación de resultados estadísticos: Permite mostrar intervalos de confianza, desviaciones estándar u otros indicadores de incertidumbre en los resultados estadísticos.

Ventajas del errorbar plot:
- Proporciona una representación visual clara de la incertidumbre o variabilidad asociada a los datos.
- Permite comparar rápidamente la dispersión entre diferentes puntos o grupos de datos.
- Ayuda a evaluar la confiabilidad de las mediciones y a interpretar los resultados de manera más completa.

Desventajas del errorbar plot:
- La interpretación precisa puede requerir conocimientos adicionales sobre los métodos de medición y los cálculos de error asociados.
- La visualización puede volverse compleja o confusa si hay muchos puntos o barras de error.

En resumen, el errorbar plot es una herramienta útil para visualizar y comprender la incertidumbre o variabilidad en los datos. Proporciona información valiosa sobre la precisión y confiabilidad de las mediciones y permite comparar la dispersión entre diferentes grupos o condiciones. Sin embargo, es importante comprender y representar adecuadamente los errores o incertidumbres asociados a los datos para una interpretación precisa y significativa.

```python
# Errorbar plot
# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

# Valores de error variables de ejemplo
yerr = 0.1 + 0.2*np.sqrt(x)
xerr = 0.1 + yerr

# Ejemplo básico utilizando la interfaz de pyplot
plt.figure()
plt.errorbar(x, y, xerr=0.2, yerr=0.4)
plt.title("Simplest errorbars, 0.2 in x, 0.4 in y")

# Cambiar a una interfaz orientada a objetos (OO) para ejercitar más características.
fig, axs = plt.subplots(nrows=2, ncols=2, sharex=True)
ax = axs[0,0]
ax.errorbar(x, y, yerr=yerr, fmt='o')
ax.set_title('Vert. symmetric')

# Reducir el número de marcas de los ejes en 4 subgráficos para evitar amontonamiento.
ax.locator_params(nbins=4)

ax = axs[0,1]
ax.errorbar(x, y, xerr=xerr, fmt='o')
ax.set_title('Hor. symmetric')

ax = axs[1,0]
ax.errorbar(x, y, yerr=[yerr, 2*yerr], xerr=[xerr, 2*xerr], fmt='--o')
ax.set_title('H, V asymmetric')

ax = axs[1,1]
ax.set_yscale('log')
# Aquí debemos asegurarnos de mantener todos los valores de y positivos:
ylower = np.maximum(1e-2, y - yerr)
yerr_lower = y - ylower

ax.errorbar(x, y, yerr=[yerr_lower, 2*yerr], xerr=xerr,
            fmt='o', ecolor='g', capthick=2)
ax.set_title('Mixed sym., log y')

fig.suptitle('Variable errorbars')

plt.show()
```

#### Network graph

Un network graph, también conocido como gráfico de redes o grafo, es una visualización que representa las relaciones y conexiones entre diferentes elementos o nodos. Consiste en nodos (representados por puntos o círculos) y arcos o líneas que conectan los nodos para mostrar las relaciones entre ellos.

El network graph se utiliza para visualizar y analizar redes complejas de conexiones y relaciones. Algunos casos de uso comunes son:

1. Redes sociales: Permite representar las conexiones entre individuos en una red social, como amistades, seguidores o interacciones.

2. Análisis de redes empresariales: Es útil para visualizar las relaciones entre diferentes empresas, departamentos o empleados en una organización.

3. Modelado de redes de transporte o infraestructura: Ayuda a representar las conexiones entre ciudades, aeropuertos, carreteras u otras infraestructuras.

Ventajas del network graph:
- Proporciona una representación visual intuitiva de las relaciones y conexiones entre diferentes elementos o nodos.
- Permite identificar rápidamente patrones, agrupaciones o nodos clave en una red compleja.
- Es útil para analizar y comprender la estructura y la dinámica de una red.

Desventajas del network graph:
- Puede volverse visualmente complejo y difícil de interpretar si hay muchos nodos o conexiones.
- La interpretación precisa puede requerir análisis más detallados y técnicas adicionales, como medidas de centralidad o comunidades.

En resumen, el network graph es una herramienta efectiva para visualizar y analizar redes complejas de conexiones y relaciones. Proporciona una representación visual clara de las interacciones entre nodos y permite identificar patrones y características clave en la red. Sin embargo, se debe tener cuidado al seleccionar y representar adecuadamente los nodos y las conexiones para garantizar una interpretación precisa y significativa del gráfico de redes.

```python
# Network graph
# Importar las librerías necesarias
import plotly.graph_objects as go
import networkx as nx

# Crear un grafo aleatorio
G = nx.random_geometric_graph(200, 0.125)

# Crear listas para las coordenadas de las aristas
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = G.nodes[edge[0]]['pos']
    x1, y1 = G.nodes[edge[1]]['pos']
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)

# Crear la traza de las aristas
edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.5, color='#888'),
    hoverinfo='none',
    mode='lines')

# Crear listas para las coordenadas de los nodos
node_x = []
node_y = []
for node in G.nodes():
    x, y = G.nodes[node]['pos']
    node_x.append(x)
    node_y.append(y)

# Crear la traza de los nodos
node_trace = go.Scatter(
    x=node_x, y=node_y,
    mode='markers',
    hoverinfo='text',
    marker=dict(
        showscale=True,
        colorscale='YlGnBu',
        reversescale=True,
        color=[],
        size=10,
        colorbar=dict(
            thickness=15,
            title='Node Connections',
            xanchor='left',
            titleside='right'
        ),
        line_width=2))

# Calcular el número de conexiones de cada nodo
node_adjacencies = []
node_text = []
for node, adjacencies in enumerate(G.adjacency()):
    node_adjacencies.append(len(adjacencies[1]))
    node_text.append('# of connections: '+str(len(adjacencies[1])))

# Asignar los valores calculados a la traza de los nodos
node_trace.marker.color = node_adjacencies
node_trace.text = node_text

# Crear la figura con las trazas de las aristas y los nodos
fig = go.Figure(data=[edge_trace, node_trace],
             layout=go.Layout(
                title='<br>Network graph made with Python',
                titlefont_size=16,
                showlegend=False,
                hovermode='closest',
                margin=dict(b=20,l=5,r=5,t=40),
                xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                )

# Mostrar la figura
fig.show()
```

#### Nightingale rose

Un Nightingale rose, también conocido como gráfico de rosa de Nightingale o gráfico polar, es un tipo de visualización que muestra la distribución y la proporción de diferentes categorías en función de un ángulo y una escala radial. Se basa en un gráfico circular dividido en segmentos o sectores, donde el ángulo de cada sector representa una categoría y la distancia desde el centro hasta el borde del gráfico indica la magnitud o la proporción de esa categoría.

El Nightingale rose se utiliza para resaltar y comparar la distribución de datos en diferentes categorías. Fue popularizado por Florence Nightingale, quien lo utilizó para visualizar la mortalidad y las causas de la muerte en los soldados durante la Guerra de Crimea. Algunos casos de uso comunes son:

1. Distribución por categoría: Permite visualizar la distribución proporcional de diferentes categorías en un conjunto de datos.

2. Comparación de proporciones: Es útil para comparar rápidamente la magnitud o la proporción de diferentes categorías y resaltar las diferencias.

3. Visualización de datos circulares: Ayuda a mostrar datos que están relacionados con un ciclo o una secuencia circular.

Ventajas del Nightingale rose:
- Proporciona una representación visual efectiva de la distribución proporcional de diferentes categorías.
- Permite una comparación rápida y clara de las magnitudes o proporciones entre categorías.
- Es útil para resaltar las categorías principales y las diferencias significativas en la distribución de datos.

Desventajas del Nightingale rose:
- La interpretación precisa puede requerir una escala clara y una comprensión adecuada de los ángulos y las proporciones representadas.
- No es adecuado para mostrar datos que no están relacionados con un ciclo o una secuencia circular.

En resumen, el Nightingale rose es una herramienta poderosa para visualizar y analizar la distribución proporcional de diferentes categorías. Permite una comparación clara y rápida de magnitudes y proporciones, y es particularmente útil para datos relacionados con un ciclo o una secuencia circular. Sin embargo, se debe tener cuidado al seleccionar y configurar adecuadamente las categorías y las escalas para garantizar una interpretación precisa y significativa del gráfico de rosa de Nightingale.

#### OLHC

Un gráfico OLHC (Open-High-Low-Close) es una visualización utilizada en el análisis técnico de los mercados financieros, especialmente en el trading de acciones, divisas y otros instrumentos financieros. Este tipo de gráfico muestra los precios de apertura (Open), máximos (High), mínimos (Low) y precios de cierre (Close) para un período específico de tiempo.

Ventajas del gráfico OLHC:
1. Información completa: El gráfico OLHC proporciona una representación visual completa de la acción del precio durante un período de tiempo determinado, mostrando tanto los niveles de apertura, máximos, mínimos y cierre de cada período.
2. Identificación de patrones: Permite identificar patrones y tendencias, como la formación de velas (candles) y patrones de reversión, que pueden ayudar a los traders a tomar decisiones informadas.
3. Análisis de volatilidad: El gráfico OLHC muestra la volatilidad del precio al resaltar las diferencias entre los precios máximos y mínimos de cada período.
4. Interpretación sencilla: Es fácil de interpretar y comprender incluso para aquellos que no están familiarizados con el análisis técnico.

Desventajas del gráfico OLHC:
1. Limitado a datos de precios: El gráfico OLHC se centra únicamente en los precios de apertura, máximos, mínimos y cierre, sin considerar otros factores o indicadores fundamentales.
2. Intervalos de tiempo fijos: La interpretación de los gráficos OLHC puede variar dependiendo del intervalo de tiempo seleccionado, lo que puede llevar a diferentes conclusiones y estrategias comerciales.
3. No proporciona información sobre el volumen de negociación: El gráfico OLHC no incluye datos sobre el volumen de operaciones, que también es una métrica importante para muchos traders.

En resumen, el gráfico OLHC es una herramienta popular en el análisis técnico que muestra los precios de apertura, máximos, mínimos y cierre para un período de tiempo determinado. Proporciona una visión completa de la acción del precio y ayuda a los traders a identificar patrones, tendencias y niveles clave. Sin embargo, es importante tener en cuenta sus limitaciones y considerar otros factores, como el volumen de negociación, al tomar decisiones comerciales.

#### Radialbar plot

Un Radialbar plot, también conocido como gráfico de barras radiales, es una visualización que representa datos en forma de barras radiales alrededor de un punto central. Cada barra radial representa una variable y su longitud o tamaño indica el valor correspondiente.

Desventajas del Radialbar plot:

- La precisión de la interpretación puede verse afectada por la escala y la legibilidad del gráfico.

En resumen, el Radialbar plot es una herramienta efectiva para visualizar y comparar valores de diferentes variables en relación con un punto central. Permite una comparación rápida y resalta las diferencias y proporciones de los valores. Sin embargo, se debe tener cuidado al usarlo con un gran número de variables o valores para garantizar la legibilidad y la interpretación precisa del gráfico de barras radiales.

```python
# Radialbar plot
import matplotlib.pyplot as plt
from matplotlib import cm
from math import log10

# Etiquetas de los segmentos
labels = list("ABCDEFG")
# Datos para cada segmento
data = [21, 57, 88, 14, 76, 91, 26]
# Número de puntos de datos
n = len(data)
# Encontrar el valor máximo para el anillo completo
k = 10 ** int(log10(max(data)))
m = k * (1 + max(data) // k)

# Radio del gráfico de donut
r = 1.5
# Calcular el ancho de cada anillo
w = r / n

# Crear colores a lo largo de un mapa de colores elegido
colors = [cm.terrain(i / n) for i in range(n)]

# Crear la figura y el eje
fig, ax = plt.subplots()
ax.axis("equal")

# Crear los anillos del gráfico de donut
for i in range(n):
    # Ocultar las etiquetas en los segmentos con textprops: alpha = 0 - transparente, alpha = 1 - visible
    innerring, _ = ax.pie([m - data[i], data[i]], radius=r - i * w, startangle=90, labels=["", labels[i]], labeldistance=1 - 1 / (1.5 * (n - i)), textprops={"alpha": 0}, colors=["white", colors[i]])
    plt.setp(innerring, width=w, edgecolor="white")

plt.legend()
plt.show()
```

- Permite una comprensión rápida y una identificación visual de los temas o conceptos clave.
- Es útil para resumir y comunicar información textual de manera concisa y atractiva.

Desventajas del Wordcloud:
- La representación visual no tiene en cuenta el contexto o el significado preciso de las palabras.
- Las palabras más frecuentes pueden dominar visualmente, ocultando otras palabras menos frecuentes pero relevantes.

En resumen, el Wordcloud es una herramienta efectiva para visualizar las palabras clave o términos más frecuentes en un texto o conjunto de textos. Permite una comprensión rápida y atractiva de los temas o conceptos principales y es útil para resumir y comunicar información textual. Sin embargo, se debe tener en cuenta que la representación visual no considera el contexto y que las palabras más frecuentes pueden dominar visualmente la nube de palabras.

```python
# Wordcloud

# importar las librerias de uso
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# crear una lista de palabras
text = "Wordclouds.com is a free online word cloud generator and tag cloud creator. Wordclouds.com works on your PC, Tablet or smartphone. Paste text, upload a document or open an URL to automatically generate a word- or tag cloud. Or enter individual words manually in the word list. Pick a shape, select colors and fonts and choose how to draw the words. Wordclouds.com can also generate clickable word clouds with links (image map). When you are satisfied with the result, save the image and share it online."

# Se crea el canvas para el wordcloud
wordcloud = WordCloud(width = 480, height = 480, margin=0).generate(text)

# mostrar el mapa de palabras
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0,y=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Cargar los datos desde el archivo 'temporal.csv'
df = pd.read_csv('temporal.csv')

# Graficar las líneas de cada columna
plt.plot(df['Mes'], df['data science'], label='data science')
plt.plot(df['Mes'], df['machine learning'], label='machine learning')
plt.plot(df['Mes'], df['deep learning'], label='deep learning')

plt.xlabel('Date')
plt.ylabel('Popularity')
plt.title('Popularity of AI terms by date')
plt.grid(True)

# Agregar texto y anotación al gráfico
plt.text(0, 80, r'$\lambda=1, r^2=0.8$')
plt.annotate('Notice something?', xy=('2014-01-01', 30), xytext=('2006-01-01', 50), arrowprops={'facecolor':'red', 'shrink':0.05})

# Mostrar el gráfico
plt.legend()
plt.show()
```
