## Cálculo de promedio de una serie de números

Ingresar una serie de números y calcula el promedio de esos números:

```python
# Creamos una lista vacía para almacenar los números ingresados por el usuario
numlista = list()

# Solicitamos al usuario que ingrese números hasta que escriba 'fin'
while True:
    inp = input('Ingresa un número (o escribe "fin" para terminar): ')
    if inp == 'fin':
        break
    
    # Convertimos el valor ingresado a tipo float y lo agregamos a la lista numlista
    valor = float(inp)
    numlista.append(valor)

# Calculamos el promedio dividiendo la suma de los números entre la cantidad de números ingresados
promedio = sum(numlista) / len(numlista)

# Imprimimos el promedio en pantalla
print('Promedio:', promedio)
```

## Cálculo de promedio a partir de la suma acumulada

Calcular el promedio de una serie de números ingresados por el usuario. El código debe permitir al usuario ingresar una serie de números y luego calcula el promedio de esos números mediante la suma acumulada y el contador de elementos.

```python
# Inicializamos la variable 'total' como 0 para almacenar la suma acumulada de los números ingresados
total = 0

# Inicializamos el contador en 0 para llevar la cuenta de la cantidad de números ingresados
contador = 0

# Solicitamos al usuario que ingrese números hasta que escriba 'fin'
while True:
    inp = input('Ingresa un número (o escribe "fin" para terminar): ')
    if inp == 'fin':
        break
    
    # Convertimos el valor ingresado a tipo float y lo almacenamos en la variable 'valor'
    valor = float(inp)
    
    # Actualizamos el valor de 'total' sumándole el número ingresado
    total = total + valor
    
    # Incrementamos el contador en 1 para llevar la cuenta de los números ingresados
    contador = contador + 1

# Calculamos el promedio dividiendo la suma acumulada ('total') entre la cantidad de números ingresados ('contador')
promedio = total / contador

# Imprimimos el promedio en pantalla
print('Promedio:', promedio)
```

## Lista de la compra

El objetivo del ejercicio es crear un programa que permita al usuario realizar diferentes acciones relacionadas con una lista de la compra. El programa debe proporcionar un menú con las siguientes opciones:

- Insertar elementos en la lista de compra.
- Ordenar los elementos de la lista de compra.
- Buscar un elemento en la lista de compra.
- Eliminar un elemento de la lista de compra.
- Modificar un elemento de la lista de compra.
- Terminar el programa.

El programa debe solicitar al usuario que elija una opción del menú y ejecutar la acción correspondiente. Por ejemplo, si el usuario elige la opción 1, se le pedirá que ingrese un elemento para agregarlo a la lista de compra. Si elige la opción 2, se le pedirá que seleccione el orden (ascendente o descendente) en el que desea ordenar la lista. Si elige la opción 3, se le pedirá que ingrese un elemento para buscarlo en la lista. Si elige la opción 4, se le pedirá que ingrese un elemento para eliminarlo de la lista. Si elige la opción 5, se le permitirá modificar un elemento existente en la lista.

El programa debe repetir el ciclo hasta que el usuario elija la opción 6 para terminar el programa.

```python
def menu():
    """
    Muestra el menú de opciones y solicita al usuario una opción válida.

    Returns:
    int: Opción seleccionada por el usuario.
    """
    while True:
        # Imprime una línea en blanco para separar visualmente el menú
        print("\n")
        # Muestra el título de la tarea
        print("Tarea: manejar lista de la compra.")
        # Muestra las opciones disponibles para el usuario
        print("1- Insertar elementos en la lista de compra")
        print("2- Ordenar los elementos de la lista de compras")
        print("3- Buscar un elemento en la lista de la compra")
        print("4- Eliminar un elemento de la lista de la compra")
        print("5- Modificar un elemento de la lista de compra")
        print("6- Terminar el programa")
        # Imprime otra línea en blanco
        print("\n")
        # Solicita al usuario que introduzca su opción
        opc = input("Teclee su opción: ")
        try:
            # Intenta convertir la opción a un entero
            opc = int(opc)
            # Verifica si la opción está dentro del rango válido (1 a 6)
            if opc in range(1, 7):
                # Devuelve la opción válida
                return opc
            else:
                # Informa al usuario si la opción no es válida
                print("La opción ingresada no es válida. Intente nuevamente.")
        except ValueError:
            # Captura el error si la entrada no es un número entero
            print("La opción ingresada no es válida. Intente nuevamente.")


def insertar_elemento(lista_compra):
    """
    Permite al usuario insertar elementos en la lista de la compra.

    Args:
    lista_compra (list): Lista de la compra actual.

    Returns:
    list: Lista de la compra actualizada.
    """
    # Inicializa la variable elemento con una cadena vacía
    elemento = ""
    # Bucle que continúa hasta que el usuario introduzca "fin"
    while elemento != "fin":
        # Solicita al usuario que introduzca un elemento
        elemento = input("Dame el elemento que vas a insertar en la lista ('fin' para terminar): ")
        # Comprueba si el elemento no es "fin" y si no está ya en la lista
        if elemento != "fin" and not busca_elemento(elemento, lista_compra):
            # Añade el elemento a la lista de compra
            lista_compra.append(elemento)
    # Devuelve la lista de compra actualizada
    return lista_compra


def ordenar_lista(lista_compra):
    """
    Ordena los elementos de la lista de la compra en orden ascendente o descendente.

    Args:
    lista_compra (list): Lista de la compra actual.

    Returns:
    list: Lista de la compra ordenada.
    """
    # Muestra las opciones de ordenamiento al usuario
    print("1- Ordenar de menor a mayor")
    print("2- Ordenar de mayor a menor")
    # Imprime una línea en blanco
    print("\n")

    while True:
        # Solicita al usuario que elija el orden de la lista
        opc = input("Seleccione el orden que quiere dar a la lista: ")
        try:
            # Intenta convertir la opción a un entero
            opc = int(opc)
            # Ordena la lista de menor a mayor si se selecciona la opción 1
            if opc == 1:
                lista_compra.sort()
                break
            # Ordena la lista de mayor a menor si se selecciona la opción 2
            elif opc == 2:
                lista_compra.sort(reverse=True)
                break
            else:
                # Informa al usuario si la opción no es válida
                print("La opción ingresada no es válida. Intente nuevamente.")
        except ValueError:
            # Captura el error si la entrada no es un número entero
            print("La opción ingresada no es válida. Intente nuevamente.")

    # Devuelve la lista de compra ordenada
    return lista_compra


def busca_elemento(elemento, lista_compra):
    """
    Busca un elemento en la lista de la compra.

    Args:
    elemento (str): Elemento a buscar.
    lista_compra (list): Lista de la compra actual.

    Returns:
    bool: True si el elemento está en la lista, False de lo contrario.
    """
    # Recorre cada elemento de la lista de compra
    for elemento_lista in lista_compra:
        # Compara el elemento de la lista con el elemento buscado
        if elemento_lista == elemento:
            # Devuelve True si se encuentra el elemento
            return True
    # Devuelve False si el elemento no se encuentra en la lista
    return False


def elimina_elemento(elemento, lista_compra):
    """
    Elimina un elemento de la lista de la compra.

    Args:
    elemento (str): Elemento a eliminar.
    lista_compra (list): Lista de la compra actual.

    Returns:
    bool: True si se eliminó el elemento correctamente, False de lo contrario.

    Raises:
    ValueError: Si el elemento no se encuentra en la lista.
    """
    # Comprueba si el elemento está en la lista
    if busca_elemento(elemento, lista_compra):
        # Elimina el elemento de la lista
        lista_compra.remove(elemento)
        # Devuelve True si el elemento fue eliminado correctamente
        return True
    else:
        # Lanza una excepción si el elemento no se encuentra en la lista
        raise ValueError("No se puede eliminar un elemento que no está en la lista")


def modificar_elemento(lista_compra):
    """
    Modifica un elemento en la lista de la compra.

    Args:
    lista_compra (list): Lista de la compra actual.

    Returns:
    list: Lista de la compra actualizada.
    """
    # Solicita al usuario el elemento que desea modificar
    elemento = input("Dime qué elemento quieres modificar: ")
    # Verifica si el elemento está en la lista
    if busca_elemento(elemento, lista_compra):
        # Solicita el nuevo valor para el elemento
        nuevo_elemento = input(f"Dime el nuevo valor para <{elemento}>: ")
        # Encuentra el índice del elemento y lo actualiza con el nuevo valor
        index = lista_compra.index(elemento)
        lista_compra[index] = nuevo_elemento
        # Informa al usuario que el elemento ha sido modificado
        print(f"El elemento <{elemento}> ha sido cambiado por <{nuevo_elemento}>.")
    else:
        # Informa al usuario si el elemento no se encuentra en la lista
        print(f"El elemento <{elemento}> no se encuentra en la lista.")
    
    # Devuelve la lista de compra actualizada
    return lista_compra


# Crea una lista vacía para almacenar los elementos de la compra
lista_compra = list()

# Llama a la función menu para mostrar el menú y obtener la opción seleccionada
opc = menu()
# Bucle que continúa hasta que el usuario seleccione la opción 6 (terminar el programa)
while opc != 6:
    # Si la opción es 1, inserta elementos en la lista de compra
    if opc == 1:
        lista_compra = insertar_elemento(lista_compra)
    # Si la opción es 2, ordena la lista de compra si tiene elementos
    elif opc == 2:
        if len(lista_compra) > 0:
            lista_compra = ordenar_lista(lista_compra)
        else:
            # Informa al usuario si la lista está vacía
            print("Primero hay que ingresar valores a la lista.")
    # Si la opción es 3, busca un elemento en la lista si tiene elementos
    elif opc == 3:
        if len(lista_compra) > 0:
            # Solicita al usuario el elemento que desea buscar
            elemento = input("Dime qué elemento quieres buscar: ")
            # Verifica si el elemento está en la lista
            encontrado = busca_elemento(elemento, lista_compra)
            if encontrado:
                # Informa al usuario si el elemento se encuentra en la lista
                print(f"El elemento <{elemento}> SI se encuentra en la lista")
            else:
                # Informa al usuario si el elemento no se encuentra en la lista
                print(f"El elemento <{elemento}> NO se encuentra en la lista")
        else:
            # Informa al usuario si la lista está vacía
            print("Primero hay que ingresar valores a la lista.")
    # Si la opción es 4, elimina un elemento de la lista si tiene elementos
    elif opc == 4:
        if len(lista_compra) > 0:
            # Solicita al usuario el elemento que desea eliminar
            elemento = input("Dime qué elemento quieres borrar: ")
            try:
                # Intenta eliminar el elemento de la lista
                elimina_elemento(elemento, lista_compra)
                # Informa al usuario si el elemento fue eliminado
                print(f"El elemento <{elemento}> SI fue borrado de la lista")
            except ValueError as error:
                # Informa al usuario si ocurre un error al eliminar el elemento
                print(str(error))
        else:
            # Informa al usuario si la lista está vacía
            print("Primero hay que ingresar valores a la lista.")
    # Si la opción es 5, modifica un elemento de la lista si tiene elementos
    elif opc == 5:
        if len(lista_compra) > 0:
            lista_compra = modificar_elemento(lista_compra)
        else:
            # Informa al usuario si la lista está vacía
            print("Primero hay que ingresar valores a la lista.")

    # Llama a la función menu nuevamente para continuar el ciclo
    opc = menu()
```

## Cálculo del promedio y guardado de datos utilizando `Pickle`

El objetivo del ejercicio es crear un programa que permita al usuario ingresar una serie de números, calcular el promedio de los números ingresados y guardar la lista de números en un archivo utilizando el módulo `pickle`.

El programa solicita al usuario que ingrese una serie de números. Si el usuario ingresa la palabra `fin`, el programa finaliza la entrada de números. Luego, calcula el promedio de los números ingresados y lo muestra en pantalla.

Posteriormente, el programa guarda la lista de números en un archivo utilizando el módulo pickle. El archivo se crea o sobrescribe en modo binario y se nombra como `iris_model.pkl`. Este archivo contiene la lista de números ingresados por el usuario.

Por último, el programa carga los datos guardados desde el archivo utilizando `pickle.load()` y los muestra en pantalla.

En resumen, el ejercicio consiste en implementar un programa que permita al usuario ingresar una serie de números, calcule el promedio de esos números, guarde la lista de números en un archivo utilizando `pickle` y luego sea capaz de cargar y mostrar los datos guardados desde el archivo.

```python
import pickle as pk  # Importar el módulo pickle y renombrarlo como pk

def solicitar_numero():
    """
    Solicita al usuario que ingrese un número o 'fin' para terminar.
    Valida la entrada y devuelve un número o None si el usuario elige terminar.
    
    Returns:
    float or None: El número ingresado o None si se ingresa 'fin'.
    """
    while True:
        inp = input("Ingresa un número (o 'fin' para terminar): ")  # Explicar claramente las opciones
        if inp.lower() == 'fin':
            return None  # Indicar que el proceso de entrada ha terminado
        try:
            return float(inp)  # Intentar convertir la entrada a un número flotante
        except ValueError:
            print("Entrada no válida. Por favor, ingresa un número.")  # Aclarar el error si la entrada es inválida

def main():
    total = 0  # Inicializar la variable total en 0
    contador = 0  # Inicializar la variable contador en 0
    lista = list()  # Crear una lista vacía para almacenar los números

    while True:
        valor = solicitar_numero()  # Usar la función para obtener un número del usuario
        if valor is None:
            break  # Salir del bucle si el usuario elige terminar

        lista.append(valor)  # Agregar el número a la lista
        total += valor  # Actualizar el total sumando el valor ingresado
        contador += 1  # Incrementar el contador en 1

    if contador > 0:
        promedio = total / contador  # Calcular el promedio dividiendo el total entre el contador
        print(f"Promedio: {promedio}")  # Imprimir el promedio calculado
    else:
        print("No se ingresaron números válidos.")  # Mensaje si no se ingresaron números

    print("Salvando la lista en un archivo...")

    filename = 'numeros_lista.pkl'  # Nombre del archivo donde se guardará la lista

    # Utilizar pickle para guardar la lista en el archivo
    # El modo 'wb' indica que se escribirá en modo binario
    with open(filename, 'wb') as f:
        pk.dump(lista, f)

    print("La lista ha sido guardada en el archivo 'numeros_lista.pkl'.")

    print("Leyendo los datos salvados desde el archivo...")

    # Cargar los datos del archivo utilizando pickle
    # El modo 'rb' indica que se leerá en modo binario
    with open(filename, 'rb') as f:
        loaded_data = pk.load(f)

    print("Datos cargados desde el archivo:")
    print(loaded_data)  # Imprimir los datos cargados desde el archivo

if __name__ == "__main__":
    main()
```

a)	Escribe una función llamada `agregar_producto` que tome como parámetros el nombre de un producto y la cantidad disponible. La función debe agregar el producto al inventario si no existe previamente, o actualizar la cantidad si el producto ya existe.

```python
def agregar_producto(inventario, nombre, cantidad):
    """
    Agrega un producto al inventario o actualiza la cantidad si el producto ya existe.

    Args:
    inventario (dict): Diccionario que representa el inventario de la tienda.
    nombre (str): Nombre del producto a agregar o actualizar.
    cantidad (int): Cantidad disponible del producto.

    Returns:
    dict: El inventario actualizado.
    """
    inventario[nombre] = cantidad
    return inventario
```

b)	Escribe una función llamada `buscar_producto` que tome como parámetro el nombre de un producto y busque si existe en el inventario. Si el producto existe, la función debe mostrar su cantidad disponible; de lo contrario, debe mostrar un mensaje indicando que el producto no se encontró.

```python
def buscar_producto(inventario, nombre):
    """
    Busca un producto en el inventario y muestra su cantidad disponible si existe.

    Args:
    inventario (dict): Diccionario que representa el inventario de la tienda.
    nombre (str): Nombre del producto a buscar.

    Returns:
    None
    """
    if nombre in inventario:
        cantidad = inventario[nombre]
        print(f"El producto '{nombre}' tiene una cantidad disponible de {cantidad}")
    else:
        print(f"El producto '{nombre}' no se encontró en el inventario")
```

c)	Escribe una función llamada `actualizar_cantidad` que tome como parámetros el nombre de un producto y una cantidad. La función debe actualizar la cantidad disponible del producto en el inventario.

```python
def actualizar_cantidad(inventario, nombre, cantidad):
    """
    Actualiza la cantidad disponible de un producto en el inventario.

    Args:
    inventario (dict): Diccionario que representa el inventario de la tienda.
    nombre (str): Nombre del producto a actualizar.
    cantidad (int): Nueva cantidad disponible del producto.

    Returns:
    dict: El inventario actualizado.
    """
    if nombre in inventario:
        inventario[nombre] = cantidad
        print(f"La cantidad del producto '{nombre}' ha sido actualizada a {cantidad}")
    else:
        print(f"El producto '{nombre}' no se encontró en el inventario")
```

d)	Escribe una función llamada `eliminar_producto` que tome como parámetro el nombre de un producto y elimine ese producto del inventario, si existe.

```python
def eliminar_producto(inventario, nombre):
    """
    Elimina un producto del inventario si existe.

    Args:
    inventario (dict): Diccionario que representa el inventario de la tienda.
    nombre (str): Nombre del producto a eliminar.

    Returns:
    dict: El inventario actualizado.
    """
    if nombre in inventario:
        del inventario[nombre]
        print(f"El producto '{nombre}' ha sido eliminado del inventario")
    else:
        print(f"El producto '{nombre}' no se encontró en el inventario")
```

Programa principal

```python
inventario = {}

while True:
    print("\n")
    print("1- Agregar un producto al inventario")
    print("2- Buscar un producto en el inventario")
    print("3- Actualizar la cantidad de un producto en el inventario")
    print("4- Eliminar un producto del inventario")
    print("5- Salir")
    print("\n")
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        nombre = input("Ingrese el nombre del producto: ")
        cantidad = int(input("Ingrese la cantidad disponible del producto: "))
        inventario = agregar_producto(inventario, nombre, cantidad)
        print(f"El producto '{nombre}' ha sido agregado al inventario")

    elif opcion == "2":
        nombre = input("Ingrese el nombre del producto a buscar: ")
        buscar_producto(inventario, nombre)

    elif opcion == "3":
        nombre = input("Ingrese el nombre del producto a actualizar: ")
        if nombre in inventario:
            cantidad = int(input("Ingrese la nueva cantidad disponible: "))
            actualizar_cantidad(inventario, nombre, cantidad)
        else:
            print(f"El producto '{nombre}' no se encontró en el inventario")

    elif opcion == "4":
        nombre = input("Ingrese el nombre del producto a eliminar: ")
        eliminar_producto(inventario, nombre)

    elif opcion == "5":
        print("¡Hasta luego!")
        break

    else:
        print("Opción inválida. Por favor, seleccione una opción válida.")
```
