### Métricas de Evaluación

Para determinar cuál de los clasificadores es más eficaz, se emplean métricas de evaluación comunes en clasificación binaria, incluyendo:

1. `Matriz de Confusión`: Un resumen de verdaderos positivos, verdaderos negativos, falsos positivos y falsos negativos.
2. `Accuracy`: tasa de clasificacion global.
3. `Precisión`: La tasa de muestras clasificadas correctamente.
4. `Recall`: La tasa de verdaderos positivos en relación con positivos reales.
5. `F1-Score`: La media armónica entre precisión y recall, buscando equilibrar ambas métricas.
6. `Curva ROC` (Receiver Operating Characteristic): Representa la relación entre tasas de verdaderos positivos y falsos positivos en diferentes umbrales de decisión.
7. `Área bajo la Curva ROC` (AUC-ROC): Una medida de la capacidad de discriminación del clasificador.

### Validación cruzada

La validación cruzada, en el contexto del aprendizaje automático y la evaluación de modelos, es una técnica fundamental que se utiliza para medir la capacidad de generalización de un modelo. Su objetivo principal es estimar cómo se comportará un modelo en datos no vistos, es decir, en datos que no se utilizaron durante su entrenamiento.

La validación cruzada aborda un problema común en el aprendizaje automático: el sobreajuste o el subajuste de un modelo. El sobreajuste ocurre cuando un modelo se ajusta demasiado a los datos de entrenamiento y no puede generalizar bien a nuevos datos, lo que resulta en un mal rendimiento en el mundo real. Por otro lado, el subajuste ocurre cuando un modelo es demasiado simple para capturar las complejidades de los datos de entrenamiento.

Para abordar estos problemas, la validación cruzada divide el conjunto de datos en dos partes: un conjunto de entrenamiento y un conjunto de prueba. El modelo se entrena en el conjunto de entrenamiento y se evalúa en el conjunto de prueba. Esto proporciona una estimación inicial del rendimiento del modelo.

Sin embargo, una única división de datos en entrenamiento y prueba puede llevar a estimaciones sesgadas, ya que depende de una única muestra de datos. La validación cruzada resuelve este problema dividiendo los datos en múltiples conjuntos de entrenamiento y prueba, realizando múltiples evaluaciones. Los resultados se promedian para obtener una estimación más precisa del rendimiento del modelo.

El método más común de validación cruzada es la validación cruzada k-fold. En este enfoque, los datos se dividen en k partes (pliegues) de igual tamaño. El modelo se entrena k veces, utilizando un pliegue diferente como conjunto de prueba en cada iteración, mientras que el resto de los pliegues se utilizan como conjunto de entrenamiento. Esto garantiza que cada punto de datos se utilice tanto para entrenamiento como para prueba, evitando el sesgo.

La validación cruzada es una herramienta esencial para la selección de modelos, la afinación de hiperparámetros y la evaluación precisa del rendimiento de un modelo. Ayuda a los científicos de datos y los ingenieros de aprendizaje automático a construir modelos más robustos y precisos que puedan generalizar bien a nuevos datos.

```python
results = {}

for classifier_name, classifier in classifiers:
    results[classifier_name] = {}
    for metric_name, metric_func in metrics.items():
        scores = cross_val_score(classifier, X, y, cv=10, scoring=metrics[metric_name])
        results[classifier_name][metric_name] = scores
```
